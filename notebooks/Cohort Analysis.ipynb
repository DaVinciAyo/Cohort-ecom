# COHORT ANALYSIS

Understanding Cohort Analysis: A Powerful Tool for Customer Retention
Cohort analysis is a critical technique in customer retention and behavioral analytics, enabling businesses to track groups of customers (cohorts) based on shared characteristics like their first purchase date. By analyzing these cohorts over time, companies can uncover trends in customer engagement, retention rates, and churn behavior, allowing them to optimize their strategies and improve long-term loyalty.


```python
#import libs
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
datatrend=pd.read_csv(r"C:\Users\DELL\Downloads\cleaned ecommence_data.csv")
```

```python
datatrend["InvoiceDate"] = pd.to_datetime(datatrend["InvoiceDate"], errors="coerce")
```

```python
datatrend.head()
```

```python
def get_cohort_date(datatrend):
    datatrend["cohort date"] = datatrend.groupby("CustomerID")["InvoiceDate"].transform("min")
    return datatrend  # Return full DataFrame
```

```python
 datatrend = get_cohort_date(datatrend) 
```

```python
# Convert InvoiceDate to datetime format
datatrend['InvoiceDate'] = pd.to_datetime(datatrend['InvoiceDate'], errors='coerce')

# Now apply the cohort grouping
datatrend['CohortMonth'] = datatrend.groupby('CustomerID')['InvoiceDate'].transform('min').dt.to_period('M')

print(datatrend[['CustomerID', 'InvoiceDate', 'CohortMonth']].head())
```

```python
datatrend.head()
```

```python
def get_year_and_month(datatrend, col):
    month = datatrend[col].dt.month
    year = datatrend[col].dt.year
    return month, year
```

```python
datatrend['cohort date'] = pd.to_datetime(datatrend['cohort date'], errors='coerce')

# Extract year and month
first_month = datatrend['cohort date'].dt.month
first_year = datatrend['cohort date'].dt.year
```

```python
latest_month = datatrend['InvoiceDate'].dt.month
latest_year = datatrend['InvoiceDate'].dt.year
```

```python
def create_cohort_index(first_month, first_year, latest_month,latest_year):
    year_diff = latest_year - first_year
    month_diff = latest_month - first_month
    index = year_diff*12 + month_diff +1 # +1 is added because of customers who have been active for just 1 month
    return index
```

```python
datatrend["cohort_index"] = create_cohort_index(first_month, first_year, latest_month, latest_year)
```

```python
datatrend
```

```python
cohort_info = datatrend.groupby(["cohort date","cohort_index"])["CustomerID"].nunique().reset_index()
cohort_info.rename(columns = {"CustomerID": "Number of customers"}, inplace = True)
cohort_info
```

```python
# create a pivot table
cohort_table = cohort_info.pivot(index = "cohort date", columns = ["cohort_index"], values = "Number of customers")

#change index to understandable format
cohort_table.index = cohort_table.index.strftime('%B %Y')
cohort_table
```

```python
plt.figure(figsize = (21,10))
sns.heatmap(cohort_table, annot = True, cmap = 'Dark2_r',  fmt = '.2f')
```

```python
# showing retention rate in percentage
new_cohort_table = cohort_table.divide(cohort_table.iloc[:,0], axis = 0)
new_cohort_table
```

```python
plt.figure(figsize = (21,10))
sns.heatmap(new_cohort_table, annot = True, cmap = 'Dark2_r', fmt = '.0%')
```

**cohort analysis heatmap** presents a clear view of customer retention trends across different months from **December 2010 to December 2011**. Here’s what stands out:

### **Key Observations**
- **Strong initial retention**: Every cohort starts at 100% in their first month.
- **Declining retention rates**: A gradual drop-off occurs across the months, indicating customer churn.
- **Best-performing cohorts**: December 2010 maintains solid retention up to **Month 12**, with a notable **50% retention rate** in its final month.
- **Drop in later cohorts**: Retention rates after September 2011 show a **steeper decline**, meaning customers acquired later were less engaged long-term.

### **Possible Insights & Next Steps**
- **Why does retention drop?** Investigate whether seasonality, product offerings, or customer engagement strategies influenced lower retention.
- **Comparing early vs. late cohorts**: Understanding differences in retention between **December 2010 and November 2011** could provide actionable insights.
- **Predicting future churn trends**: Given your experience with **customer retention analysis**, machine learning techniques like **survival analysis or time-series modeling** could forecast future retention patterns.

- **Fluctuations in retention rates**: Some cohorts show small increases or stability in certain months, suggesting seasonal effects or engagement strategies.
- **Older cohorts (early 2011) seem to retain better**: Cohorts from December 2010 and early 2011 have slightly higher retention rates in later months compared to newer cohorts.

# Churn Timing Analysis

```python

active_months = datatrend.groupby(['CustomerID'])['cohort_index'].nunique()

# Basic churn stats
average_lifetime = active_months.mean()
median_lifetime = active_months.median()

print(f"Average customer lifetime: {average_lifetime:.2f} months")
print(f"Median customer lifetime: {median_lifetime:.0f} months")
```

```python
# A customer is churned if they haven't purchased again for 3 months after their last transaction
churn_threshold = 3
last_purchase = datatrend.groupby('CustomerID')['InvoiceDate'].max()

cutoff_date = datatrend['InvoiceDate'].max() - pd.DateOffset(months=3)
churn_flag = last_purchase < cutoff_date

customer_churn_df = pd.DataFrame({
    'Last time Purchase': last_purchase,
    'Churned': churn_flag
})

print(customer_churn_df.head())
```

```python
# Count churned vs non-churned customers
churn_counts = customer_churn_df['Churned'].value_counts()

# Create a bar chart
plt.figure(figsize=(7, 5))
churn_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Customer Churn Distribution')
plt.xticks(ticks=[0, 1], labels=['Not Churned', 'Churned'], rotation=0)
plt.ylabel('Number of Customers')
plt.xlabel('Churn Status')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
churn_counts.head()
```

```python
# For each customer, find how many months they were active
customer_lifetime = datatrend.groupby('CustomerID')['cohort_index'].nunique()

# Add back cohort month
cohort_info = datatrend.groupby('CustomerID')['CohortMonth'].first()
lifetime_data = pd.DataFrame({
    'CohortMonth': cohort_info,
    'Lifetime': customer_lifetime
})

# Calculate average lifetime per cohort
cohort_lifetime = lifetime_data.groupby('CohortMonth')['Lifetime'].mean()

# Plot average churn timing across cohorts
cohort_lifetime.plot(kind='bar', figsize=(10, 5), title='Average Customer Lifetime by Cohort')
plt.ylabel("Months Active Before Churn")
plt.xlabel("Cohort Month")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**"Average Customer Lifetime by Cohort"** bar chart offers valuable insights into **how long customers remain active before churning** based on their **first purchase month (cohort month)**.

### **Key Takeaways from the Chart**
- **December 2010 cohort has the longest average customer lifetime**, exceeding **5 months before churn**.
- **A steady decline follows**—later cohorts from **mid-2011** tend to have **shorter lifetimes**, averaging around **2 months**.
- **December 2011 cohort has the lowest retention**, with an average lifetime near **1 month**.

### **Potential Insights**
- **Early cohorts (late 2010 & early 2011) had stronger retention**, possibly due to better onboarding, customer engagement, or market conditions.
- **Newer cohorts (late 2011) show faster churn**, which could indicate changes in customer behavior, pricing, or competitive dynamics.
- **If this trend continues**, understanding **why customer lifetimes are decreasing** could be key to improving retention strategies.
